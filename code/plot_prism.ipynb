```python
prism = vedo.Box(
    pos=(0, 0, 0),
    length=1.0,
    width=2.0,
    height=3.0,
    #size=(),
    c='g4',
    alpha=1.0)

vedo.show(prism, axes=1, zoom=1.1, viewup='z')
```
